```python
import pdfplumber
import os
from docx import Document

def pdf_to_word(input_pdf_path, output_docx_path):
    # Load PDF file
    with pdfplumber.open(input_pdf_path) as pdf:
        # Create a new Word document
        doc = Document()

        # Iterate through each page
        for page in pdf.pages:
            # Extract text from the page
            text = page.extract_text()

            if text:
                # Split text into paragraphs
                paragraphs = text.split('\n')

                # Add each paragraph to the Word document
                for paragraph in paragraphs:
                    doc.add_paragraph(paragraph)

        # Save the Word document
        doc.save(output_docx_path)

# Example usage


input_pdf_path = 'C:/Users/rares/Downloads/Rares Ficiu CV UK.pdf'
output_docx_path = 'C:/Users/rares/Downloads/test_doc.docx'
pdf_to_word(input_pdf_path, output_docx_path)
```

```python
import os
from pathlib import Path
from pdf2docx import Converter

def convert_pdfs_to_docx(input_dir, output_dir):
    input_path = Path(input_dir)
    output_path = Path(output_dir)

    # Create the output directory if it doesn't exist
    output_path.mkdir(parents=True, exist_ok=True)

    # Iterate through all files in the input directory
    for file in input_path.glob('*.pdf'):
        # Check if the file is a PDF
        if file.suffix.lower() == '.pdf':
            output_docx_filename = f'{file.stem}.docx'
            output_docx_path = output_path / output_docx_filename

            # Convert the PDF to a Word document
            print(f'Converting {file} to {output_docx_path}')
            cv = Converter(str(file))
            cv.convert(str(output_docx_path))
            cv.close()


# Example usage
input_dir = 'C:/Users/rares/Downloads/'
output_dir = os.path.join(input_dir, 'test_folder')

convert_pdfs_to_docx(input_dir, output_dir)
```

```python
import os
import pdfplumber
from pathlib import Path
from pdf2docx import Converter
from docx import Document

def convert_pdfs_to_docx_and_text(input_dir, output_docx_dir, output_text_dir):
    input_path = Path(input_dir)
    output_docx_path = Path(output_docx_dir)
    output_text_path = Path(output_text_dir)

    # Create the output directories if they don't exist
    output_docx_path.mkdir(parents=True, exist_ok=True)
    output_text_path.mkdir(parents=True, exist_ok=True)

    # Iterate through all files in the input directory
    for file in input_path.glob('*.pdf'):
        # Check if the file is a PDF
        if file.suffix.lower() == '.pdf':
            output_docx_filename = f'{file.stem}.docx'
            output_docx_file_path = output_docx_path / output_docx_filename

            output_text_filename = f'{file.stem}.txt'
            output_text_file_path = output_text_path / output_text_filename

            # Convert the PDF to a Word document
            print(f'Converting {file} to {output_docx_file_path}')
            cv = Converter(str(file))
            cv.convert(str(output_docx_file_path))
            cv.close()

            # Extract text from the PDF and save it as a text file
            print(f'Extracting text from {file} to {output_text_file_path}')
            with pdfplumber.open(file) as pdf:
                text = ""
                for page in pdf.pages:
                    page_text = page.extract_text()
                    if page_text:
                        text += page_text + "\n"

            with open(output_text_file_path, "w") as f:
                f.write(text)

# Example usage
input_dir = 'C:/Users/rares/Downloads/'

output_docx_dir = os.path.join(input_dir, 'docx_folder')
output_text_dir = os.path.join(input_dir, 'text_folder')

convert_pdfs_to_docx_and_text(input_dir, output_docx_dir, output_text_dir)
```
